```python
# Introduction to Loops in Python:

# Loops are used to execute a block of code repeatedly until a certain condition is met. In Python, we have two types of loops:
# 1. for loop
#   - Iterating over a range
#   - Iterating over a list
#   - Iterating over a string
# 2. while loop
# 3. Loop Control Statements:
#   - break statement
#   - continue statement
#   - pass statement
# 4. Nested Loops
```

```python
range(0, 5)
```

```python
# 1. for loop:

# The for loop is used to iterate over a sequence (such as a list, tuple, string, or range) and execute a block of code for each item in the sequence.
# Iterating over a range:
for i in range(0, 5):
    print(i)

# Index starts with 0
```

```python
range(0, 10, 2)
```

```python
for i in range(0, 10, 2):
    print(i)

# range(0, 10, 2) Means 
```

```python
# Iterating over a String:

name = "Punit Middha"

for i in name:
    print(i)
```

```python
# While Loop:

# The While Loop continues to execute as long as the Condition is True

n = 0

while n<5:
    print(n)
    n = n+1
```

```python
count = 5

while count>0:
    print(count)
    count = count - 1
```

```python
# Loop Control Statements

# Break
# The break statement exits the loop prematurely

# Break Statement

for i in range(10):
    if i == 6:
        break
    print(i)
```

```python
# Continue

# The continue statement skips the current iteration and continues with the next

for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
# Pass

# The pass statement statement is a null operation, it does nothing

for i in range(5):
    if i==3:
        pass
    print(i)
```

```python
# Nested Loops
# A loop inside a loop

for i in range(5):
    for j in range(3):
        print(f"i: {i} anf j: {j}")
```

```python
# Example

# Sum of N natural numbers

n = 10
sum = 0
count = 1

while count<=n:
    sum += count
    count += 1

print("Sum of first 10 natural Numbers: ", sum)
```

```python
# Example
# Prime Numbers between 1 and 101

for num in range(2, 101): 
    is_prime = True

    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num, end=" ")
```
