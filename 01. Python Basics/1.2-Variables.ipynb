```python
# Naming conventions:
# - Variable names should be descriptive and meaningful.
# - Variable names should be in lowercase with words separated by underscores (snake_case).
# - Variable names should not start with a number.
# - Variable names should not contain spaces or special characters (except underscores).
# - case-sensitive: 'name' and 'Name' are different variables.
```

```python
# Valid variable names:
first_name = "John"
last_name = "Doe"
age = 30
is_student = True

# Invalid variable names:
# 1name = "John"  # starts with a number
# first name = "John"  # contains space
# first-name = "John"  # contains special character '-'
# first$name = "John"  # contains special character '$'
# @name = "John"  # contains special character '@'

# Printing variables:
print(first_name)
print(last_name)
print(age)
print(is_student)
```

```python
# Case sensitivity:
Name = "Alice"
name = "Bob"

print(Name)  # Output: Alice
print(name)  # Output: Bob
```

```python
# Understanding the Variables Types in Python
# 1. Integer (int): Represents whole numbers without a decimal point.
age = 25
# 2. Float (float): Represents numbers with a decimal point.
pi = 3.14
# 3. String (str): Represents a sequence of characters enclosed in quotes.
name = "Punit"
# 4. Boolean (bool): Represents a value that can be either True or False.
is_student = True
# 5. List (list): Represents an ordered collection of items, which can be of different types.
fruits = ["apple", "banana", "cherry"]
# 6. Tuple (tuple): Represents an ordered collection of items, which can be of different types, but is immutable (cannot be changed).
coordinates = (10, 20)
# 7. Dictionary (dict): Represents a collection of key-value pairs, where each key is unique.
person = {"name": "Alice", "age": 30, "city": "New York"}
# 8. Set (set): Represents an unordered collection of unique items.
unique_numbers = {1, 2, 3, 4, 5}
# 9. NoneType (None): Represents the absence of a value or a null value.
result = None
# 10. Complex (complex): Represents complex numbers with a real and an imaginary part.
complex_number = 2 + 3j
# 11. Bytes (bytes): Represents a sequence of bytes, which is an immutable type.
data = b"Hello, World!"
# 12. Bytearray (bytearray): Represents a mutable sequence of bytes.
mutable_data = bytearray(b"Hello, World!")
# 13. Memoryview (memoryview): Represents a view of memory that can be used to access the internal data of an object without copying it.
memory_view = memoryview(b"Hello, World!")
# 14. Range (range): Represents an immutable sequence of numbers, commonly used for looping a specific number of times in for loops.
numbers = range(1, 10)
# 15. Frozenset (frozenset): Represents an immutable version of a set, which cannot be modified after creation.
immutable_set = frozenset({1, 2, 3, 4, 5})


# Printing the variables and types of variables:
print("Age:", age, "Type:", type(age))
print("Pi:", pi, "Type:", type(pi))
print("Name:", name, "Type:", type(name))
print("Is Student:", is_student, "Type:", type(is_student))
print("Fruits:", fruits, "Type:", type(fruits))
print("Coordinates:", coordinates, "Type:", type(coordinates))
print("Person:", person, "Type:", type(person))
print("Unique Numbers:", unique_numbers, "Type:", type(unique_numbers))
print("Result:", result, "Type:", type(result))
print("Complex Number:", complex_number, "Type:", type(complex_number))
print("Data:", data, "Type:", type(data))
print("Mutable Data:", mutable_data, "Type:", type(mutable_data))
print("Memory View:", memory_view, "Type:", type(memory_view))
print("Numbers:", numbers, "Type:", type(numbers))
print("Immutable Set:", immutable_set, "Type:", type(immutable_set))
```

```python
# Type Checking and Type Conversion:
# Type checking:
age = 25
print(type(age))  # Output: <class 'int'>

age_string = str(age)
print(age_string)  # Output: '25'
print(type(age_string))  # Output: <class 'str'>
```

```python
name = "Punit"
print(type(name))  # Output: <class 'str'>

int(name)  # This will raise a ValueError since 'Punit' cannot be converted to an integer
```

```python
height = 5.9
print(type(height))  # Output: <class 'float'>

int(height)  # This will convert the float to an integer, resulting in 5
print(int(height))  # Output: 5
```

```python
# Dynamic Typing:
# Python is a dynamically typed language, which means that the type of a variable can change at runtime. 
# You can assign a value of one type to a variable and later assign a value of a different type to the same variable without any issues.
x = 10
print(x, type(x))  # Output: 10 <class 'int'>
x = "Now I'm a string"
print(x, type(x))  # Output: Now I'm a string <class 'str'>
```

```python
# Simple Calculator:

num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

sum_result = num1 + num2
print("The sum of", num1, "and", num2, "is:", sum_result)
sub_result = num1 - num2
print("The difference of", num1, "and", num2, "is:", sub_result)
mply_result = num1 * num2
print("The product of", num1, "and", num2, "is:", mply_result)
div_result = num1 / num2
print("The quotient of", num1, "and", num2, "is:", div_result)
floor_div_result = num1 // num2
print("The floor quotient of", num1, "and", num2, "is:", floor_div_result)
```
