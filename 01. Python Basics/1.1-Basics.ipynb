```python
# Line continuation character is used to indicate that a statement continues on the next line.
total = 1 + 2 + 3 + \
        4 + 5 + 6
print("Total is", total)
```

```python
# Multiple Statements in a Single Line
# You can write multiple statements in a single line by separating them with a semicolon (;).

x = 10; y = 20; print("x =", x); print("y =", y)
```

```python
# Understand Semantics of Python
# variable assignment
age = 25 # age is an Integer
type(age) # Output: <class 'int'>
```

```python
name = "Punit"
type(name) # Output: <class 'str'>
```

```python
# Type Inference in Python
x = 10 # x is an Integer
print(x) # Output: 10
print(type(x)) # Output: <class 'int'>
x = "Hello" # x is now a String
print(x) # Output: Hello
print(type(x)) # Output: <class 'str'>
```

```python
# IndentationError
age = 30
if age >= 18:
print("You are an adult.") # This will raise an IndentationError because the print statement is not indented.
```
