```python
# To reuse the piece of code in multiple places, we use functions and give the dynamic values in parameters

def even_or_odd(num):
    """This funtion finds even or odd"""
    if num%2 == 0:
        print("The number is even")
    else:
        print("The number is odd")

# Calling the Function
even_or_odd(25)
```

```python
# Positional arguments

# num1, num2, num3, num4 num5 -> *args

def print_numbers(*args):
    for number in args:
        print(number, end=" ")

print_numbers(1,2,3,4,5,6,"Punit")
```

```python
# Keyword arguments

# In Keyword arguments, all the parameters are in key value pairs

def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_details(name="Punit", age=25, country="India")
```

```python
# Combining Both

def print_details(*args, **kwargs):
    for number in args:
        print("Positional Arguments:", number)
    print("------------------------------")
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_details(1,2,3,4,"Punit",name="Punit", age=25, country="India")
```
