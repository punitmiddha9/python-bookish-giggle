```python
# Definition of Datatypes:
# In Python, there are several built-in data types that are commonly used to represent different kinds of data. 
# Here are some of the most common data types in Python:
# 1. Numeric Types:
#    - int: Represents integer values (e.g., 1, -5, 0).
#    - float: Represents floating-point numbers (e.g., 3.14, -0.001).
#    - complex: Represents complex numbers (e.g., 2 + 3j).
# 2. Sequence Types:
#    - list: An ordered, mutable collection of items (e.g., [1, 2, 3], ['a', 'b', 'c']).
#    - tuple: An ordered, immutable collection of items (e.g., (1, 2, 3), ('a', 'b', 'c')).
#    - range: Represents a sequence of numbers (e.g., range(0, 10)).
# 3. Text Type:
#    - str: Represents a sequence of characters (e.g., "Hello, World!").
# 4. Mapping Type:
#    - dict: Represents a collection of key-value pairs (e.g., {'name': 'Alice', 'age': 30}).
# 5. Set Types:
#    - set: An unordered collection of unique items (e.g., {1, 2, 3}).
#    - frozenset: An immutable version of a set (e.g., frozenset({1, 2, 3})).
# 6. Boolean Type:
#    - bool: Represents a boolean value, either True or False.
```

```python
age = 30
print(type(age))  # Output: <class 'int'>
```

```python
# Float number:
pi = 3.14159
print(type(pi))  # Output: <class 'float'>
```

```python
# complex Number:
complex_number = 2 + 3j
print(type(complex_number))  # Output: <class 'complex'>
```

```python
# List:
my_list = [1, 2, 3, 'a', 'b', 'c']
print(type(my_list))  # Output: <class 'list'>
```

```python
# Tuple:
my_tuple = (1, 2, 3, 'a', 'b', 'c')
print(type(my_tuple))  # Output: <class 'tuple'>
```

```python
# Range:
my_range = range(0, 10)
print(type(my_range))  # Output: <class 'range'>
```

```python
# String:
my_string = "Hello, World!"
print(type(my_string))  # Output: <class 'str'>
```

```python
# Dict:
my_dict = {'name': 'Alice', 'age': 30}
print(type(my_dict))  # Output: <class 'dict'>
```

```python
# Set:
my_set = {1, 2, 3}
print(type(my_set))  # Output: <class 'set'>
```

```python
# Frozen Set:
my_frozenset = frozenset({1, 2, 3})
print(type(my_frozenset))  # Output: <class 'frozenset'
```

```python
# Boolean:
is_valid = True
print(type(is_valid))  # Output: <class 'bool'>
```
