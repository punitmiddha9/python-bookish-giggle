```python
# Introduction to Dictionaries

# Dictionaries are unordered collections of items. They store data in key-value pairs.
# Keys must be unique and immutable (e.g., strings, numbers, or tuples), while values can be of any type.

# Introduction to Dictionaries
# Creating Dictionaries
# Accessing Dictionary Elements
# Modifying Dictionary Elements
# Dictionary Methods
# Iterating Over Dictionaries
# Nested Dictionaries
# Dictionary Comprehensions
# Practical Examples and Common Errors
```

```python
# Creating Dict

empty_dict = {}
print(empty_dict)
print(type(empty_dict))
```

```python
empty_dict = dict()
empty_dict
```

```python
student = {"name": "Punit", "age": 25, "grade": "A+"}
print(student)
```

```python
# Single key is always used

student = {"name": "Punit", "age": 25, "name": "A+"}
print(student)
```

```python
# Accessing Dictionary Elements

student = {"name": "Punit", "age": 25, "grade": "A+"}
print(student)

print(student['grade'])
print(student['name'])
```

```python
# Accessing elements using get()

print(student.get("grade"))
print(student.get("name"))
print(student.get("age"))
print("--------------------")
print(student.get("last_name"))
print(student.get("last_name", "Not Available"))
```

```python
# Modifying the Dict Elements

# Dictionary are mutable, so we can add, update or delete elements

print(student)

student["age"] = 27 # Updated value
print(student)

student["address"] = "India" # Add key and value
print(student)
```

```python
# Deleting the key and value 

del student['age']

print(student)
```

```python
# Dictionary Methods

keys = student.keys()
print(keys)

values = student.values()
print(values)
```

```python
items = student.items()

print(items)
```

```python
student_copy = student
print(student)
print(student_copy)
```

```python
student["name"] = "Punit Middha"
print(student)
print(student_copy)
```

```python
# Shallow Copy in Dictionary

student = {'name': 'Punit', 'grade': 'A+', 'address': 'India'}
print(student)

student_shallow_copy = student.copy() # Shallow Copy
print(student_shallow_copy)
```

```python
student["name"] = "Punit Middha"
print(student)
print(student_shallow_copy)
```

```python
# Iterating over Dictionary
# Keys, Values, Items

# Iterate over Keys

for keys in student.keys():
    print(keys)
```

```python
# Iterate over Values

for values in student.values():
    print(values)
```

```python
# Iterate over key value pair 

for key, value in student.items():
    print(f"{key}:{value}")
```

```python
# Nested Dictionary

students = {
    "student1": {"name": "Punit", "age": 25},
    "student2": {"name": "Alice", "age": 27}
}
print(students)
```

```python
print(students['student1'])
print(students['student1']['name'])
```

```python
# Iterating over nested dictionary

for student_id, student_info in students.items():
    print(f"{student_id}: {student_info}")
    for key, value in student_info.items():
        print(f"{key}: {value}")
```

```python
# Dictionary Comprehnsion

squares = {x: x**2 for x in range(5)}
print(squares)
```

```python
# Conditional Dictionary Comprehension

evens = {x: x**2 for x in range(5) if x%2==0}
print(evens)
```

```python
## Practical Examples

## Use a dictionary to count he frequency of elements in list

numbers=[1,2,2,3,3,3,4,4,4,4]
frequency={}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```

```python
## Merge two dictionaries into one

dict1={"a":1,"b":2}
dict2={"b":3,"c":4}
merged_dict={**dict1,**dict2}
print(merged_dict)
```
