```python
# Creating a Tuple

empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst = list()
print(type(lst))
tpl = tuple()
print(type(tpl))
```

```python
num = tuple([1,2,3,4,5])
num
```

```python
mixed_tpl = (1, "Hello", 3.14, True)
print(mixed_tpl)
```

```python
# Accessing the tuple elsement:

print(num[2])
print(num[0:4])
print(num[::-1])
```

```python
# Tuple Operations:
concat_tuple = num + mixed_tpl
print(concat_tuple)
```

```python
mixed_tpl * 3
```

```python
num * 3
```

```python
# Immutabe Nature of Tuples:

lst = [1,2,3,4,5]
print(lst)

lst[1] = "Punit"
print(lst)
```

```python
num[1] = "Punit" # Tuple is immutable
```

```python
# Tuple Methods:

print(num.count(1))
print(num.index(3))
```

```python
# Packing and Unpacking Tuple:

# Packing - Pack the values in tuple
packed_tuple = 1, "Punit", 3.14
print(packed_tuple)
```

```python
# Unpacking

a,b,c = packed_tuple

print(a)
print(b)
print(c)
```

```python
# Unpacking with *

numbers = (1,2,3,4,5,6)

first, *middle, last = numbers
print(first)
print(middle)
print(last)
```

```python
# Nested Tuple:

lst = [[1,2,3,4], [5,6,7,8], [1, "Hello", 3.14, "c"]]
lst[0][0:3]
```

```python
lst = [[1,2,3,4], [5,6,7,8], (1, "Hello", 3.14, "c")]
lst[2][0:3]
```

```python
nested_tuple = ((1,2,3), ("a", "b", "c"), (True, False))
print(nested_tuple[2])
print(nested_tuple[1][2])
```

```python
# Iterating over Nested Tuple:

for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item, end=" ")
    print()
```
