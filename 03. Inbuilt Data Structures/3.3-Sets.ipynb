```python
# Sets

# Sets are a built-in data type in Python used to store collections of unique items. 
# They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elements. 
# Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union, intersection, difference, and symmetric difference.
```

```python
# Creating a Set:

my_set = {1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
empty_set = set()
print(empty_set)
```

```python
my_set = set([1,2,3,4,5,6,5,6])
print(my_set)
```

```python
# Basic Sets Operations:

# Adding Elements:

my_set.add(7)
print(my_set)
my_set.add(8)
print(my_set)
```

```python
# Removing the Elements:

my_set.remove(7)
print(my_set)
```

```python
# Removing the element which is not present in set

my_set.remove(10)
```

```python
# Discarding an element

my_set.discard(10)
```

```python
# Pop Method

removed_ele = my_set.pop()
print(removed_ele)
print(my_set)
```

```python
my_set.clear()
print(my_set)
```

```python
# Set Membership Test:

my_set = {1,2,3,4,5}
print(3 in my_set)
print(7 in my_set)
```

```python
# Mathematical Operations

set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,8,9}

# Union
union_set = set1.union(set2)
print(union_set)

# Intersection
intersection_set = set1.intersection(set2)
print(intersection_set)

set1.intersection_update(set2)
print(set1)
```

```python
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,8,9}

# Difference

print(set1.difference(set2))
```

```python
# Symmetric Difference

print(set1.symmetric_difference(set2)) # Remove all the intersect element
```

```python
# Sets Methods

set1 = {1,2,3}
set2 = {3,4,5}

# is subset
print(set1.issubset(set2))

# is superset
print(set1.issuperset(set2))
```
